```python
import pandas as pd
import numpy as np
from pathlib import Path

# Display options
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 100)

print("Libraries loaded successfully.")
```

```python
from pathlib import Path

# Find project root automatically
project_root = Path.cwd().parent

raw_tree_path = project_root / "data" / "raw" / "trees"

print(raw_tree_path)
```

```python
tree_files = sorted(raw_tree_path.glob("*.csv"))

print(f"{len(tree_files)} files found.\n")

for file in tree_files:
    print(file.name)
```

```python
tree_data = {}

for file in tree_files:
    tree_data[file.stem] = pd.read_csv(file)

print("Datasets loaded successfully.")
```

```python
summary = []

for name, df in tree_data.items():

    summary.append({
        "Dataset": name,
        "Rows": df.shape[0],
        "Columns": df.shape[1]
    })

summary_df = pd.DataFrame(summary)

summary_df
```
